#WEB SCRAPPING https://books.toscrape.com/index.html

```python
#THE DATA WE NEED TO EXTRACT FROM THE WEBSITE

pages = []
images_url = []
titles = []
price =[]
rating =[]
```

```python
no_of_pages = 1

for i in range(1, no_of_pages+1):
  url="https://books.toscrape.com/catalogue/page-{}.html".format(i)
  pages.append(url)
```

##GET THE REQUEST FOR EACH URL

```python
import requests
import bs4

for p in pages:
  data=requests.get(p)
  soup=bs4.BeautifulSoup(data.text, "xml")

soup
```

```python
print(soup.prettify())
```

## finding the titles

```python
for i in soup.findAll("h3"):
  titl=i.getText()
  titles.append(titl)

print(titles)
```

## finding the prices

for i in soup.findAll("p", class_="price_color"):
  print(i)

```python
for i in soup.findAll("p", class_="price_color"):
  pr=i.getText()
  price.append(pr)

price
```

## finding the image urls


```python
divs = soup.find_all("div",class_="image_container")
for thumbs in divs:
  tags=thumbs.find("img",class_="thumbnail")
  print(tags)
```

```python
divs = soup.find_all("div",class_="image_container")
for thumbs in divs:
  tags=thumbs.find("img",class_="thumbnail")
  link= "https://books.toscrape.com/catalogue/"+str(tags['src'])
  images_url.append(link)

images_url
```

## finding the rating

```python
!pip install attrs
```

```python

#getting the ratings
data_r = soup.select('p.star-rating')
for i in data_r:
  for k,v in i.attrs.items():
    star=v.replace("star-rating","")
    rating.append(star)
    print(star)
```

```python
print(rating)
```
